```python
y = []

for i in df2['yr_renovated']:
    if i == 0:
        y.append('never')
    elif i < 2014 and i > 1989:
        y.append('recently')
    elif i <= 1989 and i >= 1964:   
        y.append('25+ years ago')
    else:
        y.append('50+ years ago')
                    
df2['renovation'] = y
```

```python
#living space

x = []

for n, i in enumerate(df2.year_sold):
    if i == 2014:
        x.append(df2['sqft_living'][n])
    else:
        x.append(df2['sqft_living15'][n])
                 

df2['sqft_living_at_saledate'] = x
```

### Filtering outliers for the continuous data

```python
def outliers(column, threshold = 3):
    data = column[abs(column.apply(lambda x: (x - column.mean()) / column.var() ** (1/2))) > threshold]
    return data
```

```python
sqft_lot_outliers = outliers(df4['sqft_lot_at_saledate'])
price_outliers = outliers(df4['price'])
```

## Training the Linear Regression model on the Training set

```python
# sklearn
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet, SGDRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
```

### split into test and training data

```python
X = final_df.drop(columns = "price")
Y = final_df.price
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y , test_size = 0.2)
```

### Scale the x- values

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
y_train.value_counts()
```

### train the model

```python
lr_model.coef_
```

```python
coeff_df
```

```python
#Note to myself
#How to interpret coefficients of log-transformed variables:

#Interpret the coefficient as the percent increase in the dependent variable for every 1% increase 
#in the independent variable.

#For x percent increase, calculate 1.x to the power of the coefficient, subtract 1, and multiply by 100.
# Example: For every 20% increase in the independent variable, our dependent variable increases by about 
#(1.20 0.198  1) * 100 = 3.7 percent.
```

### Get prediction

```python
y_pred= lr_model.predict(X_test)
```

```python
y_pred
```

### Compare the results

```python
import matplotlib.pyplot as plt
import numpy as np
plt.plot(np.linspace(min(y_test), max(y_test)),
         np.linspace(min(y_test), max(y_test)), c='r') 
#an artificially created line, which represents the model,
# we use min(y_test) so the graph is on the right scale, beginning with the lowest test data. 
#if we put 0 it wo show a huge line with the data points at the upper right corner
plt.scatter(y_test,y_pred, s=5) 


plt.xlabel('actual observations or y_test')
plt.ylabel('data we predicted or pred_y')
```

```python
from sklearn import metrics


print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('R2:', metrics.r2_score(y_test, y_pred))
print('Adjusted R:',  1 - (1-metrics.r2_score(y_test, y_pred))*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1))
```

### Conclusion

The model seems to be a good fit. The MSE is in a good range. 
According to the R2 value the independant variables explained the target variable quite ok. 

The size of the living area as well as grade have the strongest positive impact on the housing price, 
followed by number of bathrooms and floors. Checking the graph, the prediction seems ok, the scatterplot is fairly linear. However on the lower values as well as on the higher values show fit less well, which is probably a result of my outlier problem. see below. 

It is curious to see that the number of bedrooms has a negative impact on the housing price. This would require some deeper investigation. It could be that it is impractical at some point to have too many bedrooms. Even if one
want to buy a luxury house one does not necessary need that many bedrooms. That goes along with observations I made in Tableau where the average price for houses had its peak for 7-8 bedrooms and then decreased for any number higher than 8 bedrooms.

Similiar observation for the years in which the house was built. One would assume that houses that were built recently have a higher value. Probably there is some sort of nostalgic factor that plays into buying a house. 
Checking the data in Tableau I could observe that newer houses usually got a average score for condiion as well as an average grade, whereas among the older houses there was a higher number of high graded houses in better conditions. 




#### Possibilities to improve my model:

The model could be improved through either re-adding some of the features or try another formula for cutting the outliers. (I had a problem in my outlier function that I couldn't solve and eventually had to go back to use a formula that would cut more outliers than I wanted)

I could also try to encode my categorical data differently, use a dummy variable instead of the ordinal encoding and see if this improves the model. 

If I decide to stick to ordinal encodeing, from my observations in Tableau regarding the age of the house and the condition, I would probably go back change the order of ordinal encoding, start with the newest houses = 0 and so on. 

The waterfront location surprisingly does not seem to play a big role in the price determination. (maybe because I didn't turn it into a Boolean)

```python
def models_automation(models, X_train, y_train):
    for model in models:
        model.fit(X_train, y_train)
        print(f"{model.__class__.__name__}: Train -> {model.score(X_train, y_train)}, Test -> {model.score(X_test, y_test)}")
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn = [KNeighborsRegressor()]
models_automation(knn, X_train, y_train)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = [DecisionTreeRegressor()]
models_automation(dtr, X_train, y_train)
```
